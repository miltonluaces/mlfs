# Feed Forward Neural Network Classification with Keras
### Using Tensorflow / Keras

## Imports

```python
from utilities.std_imports import *
from utilities.tf_imports import *
```

## Load data and split input/output

```python
dataset = np.loadtxt('D:/data/csv/diabetes2.csv', delimiter=',', skiprows=1)
X = dataset[:,0:8]
y = dataset[:,8]
dataset
```

## Build model

```python
model = tkm.Sequential()
model.add(tkl.Dense(12, input_dim=8, activation='relu'))
model.add(tkl.Dense(8, activation='relu'))
model.add(tkl.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Testing

```python
model.fit(X, y, epochs=150, batch_size=10)
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
